#### Lets start to boxplot the outliers

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")


numerical_columns = customer_df.select_dtypes(include=['int64', 'float64']).columns.tolist()


for column in numerical_columns:
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=customer_df, y=column)
    plt.title(f'Boxplot of {column}')
    plt.ylabel(column)
    plt.tight_layout()
    plt.show()
```

```python
from scipy.stats import zscore

# Assuming customer_df is your DataFrame

# List of numerical columns to remove outliers from
columns_to_remove_outliers = ['customer_lifetime_value', 'monthly_premium_auto', 'total_claim_amount']

# Create a copy of the DataFrame to work on
customer_df_outliers_removed = customer_df.copy()

# Calculate Z-scores for the specified columns
z_scores = zscore(customer_df_outliers_removed[columns_to_remove_outliers])

# Define a Z-score threshold (e.g., 3)
z_score_threshold = 3

# Remove rows with Z-scores greater than the threshold
outliers_removed_indices = (z_scores < z_score_threshold).all(axis=1)
customer_df_outliers_removed = customer_df_outliers_removed[outliers_removed_indices]

# Print the shape of the DataFrame before and after outlier removal
print("Original DataFrame shape:", customer_df.shape)
print("DataFrame shape after removing outliers:", customer_df_outliers_removed.shape)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming customer_df and customer_df_outliers_removed are your DataFrames

# Set seaborn style for better aesthetics
sns.set(style="whitegrid")

# List of numerical columns
numerical_columns = ['customer_lifetime_value', 'monthly_premium_auto', 'total_claim_amount']

# Create box plots for the original DataFrame and the outliers-removed DataFrame
plt.figure(figsize=(12, 8))
for i, column in enumerate(numerical_columns):
    plt.subplot(len(numerical_columns), 2, i*2 + 1)
    sns.boxplot(data=customer_df, y=column)
    plt.title(f'Boxplot of {column} (Original)')
    plt.ylabel(column)
    plt.subplot(len(numerical_columns), 2, i*2 + 2)
    sns.boxplot(data=customer_df_outliers_removed, y=column)
    plt.title(f'Boxplot of {column} (Outliers Removed)')
    plt.ylabel(column)
    plt.tight_layout()

plt.show()

```
